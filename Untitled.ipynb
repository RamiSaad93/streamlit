```python
import pandas as pd
import numpy as np

housing_df = pd.read_csv(r"C:\Users\ramya\OneDrive\Documents\GitHub\streamlit_intro\Housing.csv")

housing_df.head()
```

```python
housing_df['bedrooms'].unique()
```

```python
housing_df['mainroad'] = housing_df['mainroad'].map({'yes': 1, 'no': 0})
housing_df['guestroom'] = housing_df['guestroom'].map({'yes': 1, 'no': 0})
housing_df['basement'] = housing_df['basement'].map({'yes': 1, 'no': 0})
housing_df['hotwaterheating'] = housing_df['hotwaterheating'].map({'yes': 1, 'no': 0})
housing_df['airconditioning'] = housing_df['airconditioning'].map({'yes': 1, 'no': 0})
housing_df['prefarea'] = housing_df['prefarea'].map({'yes': 1, 'no': 0})
housing_df['furnishingstatus'] = housing_df['furnishingstatus'].map({'furnished': 2, 'semi-furnished': 1, "unfurnished":0})
```

```python
housing_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

X = housing_df.drop(columns = "price")
y = housing_df["price"]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
len_reg = LinearRegression()
len_reg.fit(X_train, y_train)
y_pred = len_reg.predict(X_test)
score = r2_score(y_test, y_pred)
print(f"for len_reg R² score: {score}")

dt = DecisionTreeRegressor(max_depth = 3)
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
score = r2_score(y_test, y_pred)
print(f"for decision tree R² score: {score}")

rf = RandomForestRegressor(n_estimators = 200, max_depth =3)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
score = r2_score(y_test, y_pred)
print(f"for random forest R² score: {score}")

knn= KNeighborsRegressor(n_neighbors = 24)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
score = r2_score(y_test, y_pred)
print(f"for knn R² score: {score}")
```

```python
import pickle

with open("best_model.pkl", "wb") as f:
    pickle.dump(len_reg, f)

with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)
```
